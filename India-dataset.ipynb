```python
import math
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Instantiate a RandomForestClassifier
rfc = RandomForestClassifier()

# Fit the RandomForestClassifier model on the training data
model = rfc.fit(x_train, y_train)

# Calculate the accuracy of the model on the training data
train_accuracy = model.score(x_train, y_train)

# Calculate the accuracy of the model on the test data
test_accuracy = model.score(x_test, y_test)

# Predict the target variable using the trained model
y_pred = model.predict(x_test)

# Calculate the Root Mean Square Error (RMSE)
MSE = np.square(np.subtract(y_test, y_pred)).mean()
RMSE = math.sqrt(MSE)

# Calculate additional evaluation metrics
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
confusion_mat = confusion_matrix(y_test, y_pred)

# Print the evaluation results
print("Train Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
print("Root Mean Square Error:", RMSE)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:\n", confusion_mat)

```

```python
#Importing the H2O Random Forest Estimator module from the H2O package.
from h2o.estimators import H2ORandomForestEstimator
```

```python
import h2o
from h2o.frame import H2OFrame

h2o.init() # initialize h2o
h2o_frame = h2o.H2OFrame(updated_data)
type(h2o_frame)
```

```python
type(h2o_frame)
```

```python
updated_data.head()
```

```python
#Returns a list of column names in the H2O frame 
h2o_frame.columns
```

```python
predictors = ['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI','AQI_Bucket']
response = "City"
```

```python
h2o_frame[response] = h2o_frame[response].asfactor()
```

```python
data_split = h2o_frame.split_frame(ratios = [0.8], seed = 1234)
data_train = data_split[0] # using 80% for training
data_test = data_split[1]  # using the rest 20% for out-of-bag evaluation
```

```python
rfc_model = H2ORandomForestEstimator(ntrees=10,
                                    max_depth=5,
                                    min_rows=10,
                                    
                                    binomial_double_trees=True)
```

```python
#Training a random forest classifier model on training data 
rfc_model.train(x=predictors,
                   y=response,
                   training_frame=data_train)
```

```python
#Calculating performance metrics in Random Forest Classifier model on test data
perf = rfc_model.model_performance(data_test)
```

```python
perf
```

```python
#Calculate the accuracy of a classification model using mean per class error
acc = 1 - perf.mean_per_class_error()
print("Accuracy:", acc)


```

### Algorithm - 2 K Nearest Neighbours (KNN)regression

#### Model 1

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Create a KNN Regression object with k=5
knn_model = KNeighborsRegressor(n_neighbors=5)

# Fit the model to the training data
knn_model.fit(x_train, y_train)

# Predict the target variable for the test data
y_pred = knn_model.predict(x_test)

# Evaluate the model using mean squared error and R-squared metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE): {:.2f}".format(mse))
print("R-squared (R2): {:.2f}".format(r2))

```

#### Model 2

```python
from sklearn.neighbors import RadiusNeighborsRegressor

# Create a Radius Neighbors Regression object with radius=1.0
rnn_model = RadiusNeighborsRegressor(radius=1.0)

# Fit the model to the training data
rnn_model.fit(x_train, y_train)

# Predict the target variable for the test data
y_pred = rnn_model.predict(x_test)

# Evaluate the model using mean squared error and R-squared metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE): {:.2f}".format(mse))
print("R-squared (R2 ): {:.2f}".format(r2))

```

### Algorithm- 3, Support Vector Machines (SVM)

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.svm import SVR
```

```python

#train SVM model
model = SVR(kernel='rbf')
model.fit(x_train, y_train)

# evaluate model performance
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error (MSE)', mse)
print('Root Mean Squared Error (RMSE)', rmse)
print('Mean Absolute Error (MAE)', mae)
print("R^2 value: ", r2)
```

```python
from keras.models import Sequential
from keras.layers import Dense

from sklearn.metrics import r2_score
```

```python

# define the model
model = Sequential()
model.add(Dense(10, input_dim=x_train.shape[1], activation='relu'))
model.add(Dense(1, activation='linear'))

# compile the model
model.compile(loss='mean_squared_error', optimizer='adam')

# fit the model to the training data
history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=0)
```

```python
# evaluate the model performance on the test data
loss = model.evaluate(x_test, y_test, verbose=0)
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error (MSE)', loss)
print("R^2 value: ", r2)
```
